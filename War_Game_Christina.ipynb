Let's start by defining the suits, ranks, and values. 

```python
import random


Suits={"Hearts","Diamonds", "Clubs", "Spades"}

Ranks={"Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King", "Ace"}

Values= {"Two":2 , "Three":3, "Four":4, "Five":5, "Six":6, "Seven":7, "Eight":8, "Nine":9,
         "Ten":10, "Jack":11, "Queen":12, "King":13, "Ace":14}
```

First we are building the card class

```python
class Card():
    
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=Values[rank]
        
    def __str__(self):
        return self.rank+" of "+ self.suit
```

Now we are going to creat our deck class

```python
class Deck():
    
    def __init__(self):
        self.allcards=[]              # Here we just define the list
        for suit in Suits:
            for rank in Ranks:
                self.allcards.append(Card(suit,rank))    # Here we just create the whole cards
    
    def shuffle(self):
        random.shuffle(self.allcards)               # This shuffles the list of the cards
        
    def deal_one(self):
        return self.allcards.pop()     # Don't forget to use return here
```

We can try to see each objecty created but I already test them. Next we are going to create the player objects

```python
class Player():
    
    def __init__(self, name):  # we allow that each player get a name
        self.name=name
        self.all_cards=[]
        
    def remove_one(self):                             # To release one card to the table 
        return self.all_cards.pop(0)                  # Note that it is arguman of .pop is 0
    
    def addcards(self,newcards):                      # To add newcards to the hand of player
        if type(newcards)==type([]):                  # to check if multiple cards being added then do this
            self.all_cards.extend(newcards)
            ''' we can not use .append bc it makes it [A,B,[newcards]], but .extend make it [A,B,newcards]
            '''
        else:
            self.all_cards.append(newcards)
            
    def __str__(self):
        
        return f"Player {self.name} has {len(self.all_cards)} cards."    # if we print each object, this will show up
```

So now it is time to create the players and split the deck between them

```python
player_one=Player("one")                      #created the first player and name it one
player_two=Player("two")                      #created the first player and name it two

new_deck=Deck()                               # creates new deck

new_deck.shuffle()                            # shuffles the new deck

for i in range(26):                           # we have 54 cards in each deck, it splits between two players
    player_one.addcards(new_deck.deal_one())
    player_two.addcards(new_deck.deal_one())
```
